```python
import numpy as np
```

```python
voting_w_coords = np.load('votenet_tf/voting_w_coords.npy')
voting_b_coords = np.load('votenet_tf/voting_b_coords.npy')
voting_w_features = np.load('votenet_tf/voting_w_features.npy')
voting_b_features = np.load('votenet_tf/voting_b_features.npy')
voting_act_coords = np.load('votenet_tf/voting_act_coords.npy')
voting_act_features = np.load('votenet_tf/voting_act_features.npy')
```

```python
proposal_w_coords = np.load('votenet_tf/proposal_w_coords.npy')
proposal_b_coords = np.load('votenet_tf/proposal_b_coords.npy')
proposal_w_features = np.load('votenet_tf/proposal_w_features.npy')
proposal_b_features = np.load('votenet_tf/proposal_b_features.npy')
proposal_act_coords = np.load('votenet_tf/proposal_act_coords.npy')
proposal_act_features = np.load('votenet_tf/proposal_act_features.npy')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(30,30))
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(15,5))
fig.suptitle('Distribution of weights, in Voting module')

data = np.reshape(voting_w_coords, (-1))
bin_size = 0.1; min_edge = min(data); max_edge = max(data)
N = int((max_edge-min_edge)/bin_size); Nplus1 = N + 1
bin_list = np.linspace(min_edge, max_edge, Nplus1)
ax1.hist(data, bin_list)
ax1.set_xlim((-0.5,0.5))
ax1.set_xlabel('Coordinates part, mean: ' + '{0:.2f}'.format(np.mean(data)) + ', sd: ' + '{0:.2f}'.format(np.std(data)))
ax1.set_ylabel('Count')

data = np.reshape(voting_w_features, (-1))
bin_size = 0.1; min_edge = min(data); max_edge = max(data)
N = int((max_edge-min_edge)/bin_size); Nplus1 = N + 1
bin_list = np.linspace(min_edge, max_edge, Nplus1)

ax2.hist(data, bin_list)
ax2.set_xlim((-0.5,0.5))
ax2.set_xlabel('Features part, mean: ' + '{0:.2f}'.format(np.mean(data)) + ', sd: ' + '{0:.2f}'.format(np.std(data)))
ax2.set_ylabel('Count')
```
